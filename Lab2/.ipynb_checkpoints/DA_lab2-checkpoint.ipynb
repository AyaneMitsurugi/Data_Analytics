```python
# Part 1 - Sampling in Python and visualising samples

# -----------------------------------------------------------------------------------------------------------------------------------------------------------------
#                                                         NORMAL DISTRIBUTION
# -----------------------------------------------------------------------------------------------------------------------------------------------------------------

# Using histograms with and without cumulative option observe how increasing variance and mean changes look of the distribution.
# Check how number of bins influences the plots.
# Consider what would make normal distribution weekly informative and what will make it informative.
```

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Dictionary
dictionary_normal = { "Normal"               : np.random.normal(size=10000),
                      "Normal (loc=2, scale=3)" : np.random.normal(loc=2, scale=3, size=10000),
                      "Normal (loc=2, scale=5)" : np.random.normal(loc=2, scale=5, size=10000),
                      "Normal (loc=3, scale=2)" : np.random.normal(loc=3, scale=2, size=10000),
                      "Normal (loc=5, scale=2)" : np.random.normal(loc=5, scale=2, size=10000),
                      "Normal (loc=1, scale=1)" : np.random.normal(loc=1, scale=1, size=10000)}

# Dictionary --> Data Frame
df_normal = pd.DataFrame.from_dict(dictionary_normal)

# Melting Data Frame
df_normal_melt = df_normal.melt()

# WITHOUT CUMMULATIVE
grid = sns.FacetGrid(data=df_normal_melt,col='variable',col_wrap=3)
grid.map(plt.hist, "value", color="steelblue", cumulative = False)
```

```python
# WITH CUMMULATIVE
grid = sns.FacetGrid(data=df_normal_melt, col='variable', col_wrap=3)
grid.map(plt.hist, "value", color="steelblue", cumulative = True)
```

```python
grid = sns.FacetGrid(data=df_normal_melt, col='variable', col_wrap=3)
grid.map(plt.hist, "value", color="steelblue", cumulative = True)
```

```python
# -----------------------------------------------------------------------------------------------------------------------------------------------------------------
#                                                        NORMAL DISTRIBUTION (with inverse logit)
# -----------------------------------------------------------------------------------------------------------------------------------------------------------------
# Transform samples from normal distribution with inverse logit.
# Consider how mean and standard deviation influence the shape of histogram and cdf.
# Is previously non-informative normal is still non-informative after transformation?
```

```python
from scipy.special import expit
```

```python
df_expit = expit(df_normal)
df_expit_melt = df_expit.melt()
```

```python
# WITHOUT CUMMULATIVE
grid = sns.FacetGrid(data=df_expit_melt,col='variable',col_wrap=3)
grid.map(plt.hist, "value", color="steelblue", cumulative = False)
```

```python
# WITH CUMMULATIVE
grid = sns.FacetGrid(data=df_expit_melt, col='variable', col_wrap=3)
grid.map(plt.hist, "value", color="steelblue", cumulative = True)
```

```python
# ----------------------------------------------------------------------------------------------------------------------------------------------------------
#                                                              POISSON DISTRIBUTIO
# ----------------------------------------------------------------------------------------------------------------------------------------------------------

# Poisson distribution.
# This distribution returns integers (give histogram location of bins instead of their numbers), depending on intensity.
# This is a good distribution for modelling countable effects.
# Observe how mean and variance influence the shape.
```

```python
# Dictionary
dictionary_poisson = { "Poisson (interval=1)" : np.random.poisson(1,100000),
                       "Poisson (interval=3)" : np.random.poisson(3,100000),
                       "Poisson (interval=5)" : np.random.poisson(5,100000),
                       "Poisson (interval=10)" : np.random.poisson(10,100000),
                       "Poisson (interval=20)" : np.random.poisson(20,100000),
                       "Poisson (interval=50)" : np.random.poisson(50,100000)}

# Dictionary --> Data Frame
df_poisson = pd.DataFrame.from_dict(dictionary_poisson)

# Melting Data Frame
df_poisson_melt = df_poisson.melt()

# Graphs
# WITHOUT CUMMULATIVE
grid = sns.FacetGrid(data=df_poisson_melt,col='variable',col_wrap=3)
grid.map(plt.hist, "value", color="steelblue", cumulative = False)
```

```python
# WITH CUMMULATIVE
grid = sns.FacetGrid(data=df_poisson_melt,col='variable',col_wrap=3)
grid.map(plt.hist, "value", color="steelblue", cumulative = True)
```

```python
# ----------------------------------------------------------------------------------------------------------------------------------------------------------
#                                                                 BETA DISTRIBUTION
# ----------------------------------------------------------------------------------------------------------------------------------------------------------

# Beta distribution.
# This is a distribution defined on [0,1] interval.
# It is useful as a prior for probability parameters (like binomial distribution).
# Traditionally it is defined with $\alpha$ and $\beta$ parameters.
# There are however different reparametrizations possible.
# Verify how beta distributions behave for their natural parametrization,
# location dispersion parametrization and mean-population size parametrization (see lecture notes)
```

```python
# Dictionary
dictionary_beta = { "Beta (alpha=1, beta=1)" : np.random.beta(1, 1, 10000),
              "Beta (alpha=1, beta=2)" : np.random.beta(1, 2, 10000),
              "Beta (alpha=1, beta=5)" : np.random.beta(1, 5, 10000),
              "Beta (alpha=2, beta=1)" : np.random.beta(2, 1, 10000),
              "Beta (alpha=5, beta=1)" : np.random.beta(5, 1, 10000),
              "Beta (alpha=5, beta=5)" : np.random.beta(5, 5, 10000),
}

# Dictionary --> Data Frame
df_beta = pd.DataFrame.from_dict(dictionary_beta)

# Melting Data Frame
df_beta_melt = df_beta.melt()

# Graphs
# WITHOUT CUMMULATIVE
grid = sns.FacetGrid(data=df_beta_melt, col='variable', col_wrap=3)
grid.map(plt.hist, "value", color="steelblue", cumulative = False)
```

```python
# WITH CUMMULATIVE
grid = sns.FacetGrid(data=df_beta_melt, col='variable', col_wrap=3)
grid.map(plt.hist, "value", color="steelblue", cumulative = True)
```

```python
# ----------------------------------------------------------------------------------------------------------------------------------------------------------
#                                                           LOG-NORMAL DISTRIBUTION
# ----------------------------------------------------------------------------------------------------------------------------------------------------------

# Log-normal distribution. This is a great distributions for positive variables, especially if scale can differ (for example salaries).
# Assumtion is that the logarithm of variable of interest is normaly distributed.
# Observe how mean and variance of normal distribution are propagating through transformation.
```

```python
# Dictionary
dictionary_log_normal = { "Log-normal" : np.random.lognormal(size=10000),
                          "Log-normal (mean=1, sigma=0.1)" : np.random.lognormal(1, 0.1, 10000),
                          "Log-normal (mean=1, sigma=0.2)" : np.random.lognormal(1, 0.2, 10000),
                          "Log-normal (mean=1, sigma=0.5)" : np.random.lognormal(1, 0.5, 10000),
                          "Log-normal (mean=2, sigma=0.1)" : np.random.lognormal(2, 0.1, 10000),
                          "Log-normal (mean=5, sigma=0.1)" : np.random.lognormal(5, 0.1, 10000),
}

# Dictionary --> Data Frame
df_log_normal = pd.DataFrame.from_dict(dictionary_log_normal)

# Melting Data Frame
df_log_normal_melt = df_log_normal.melt()

# Graphs
# WITHOUT CUMMULATIVE
grid = sns.FacetGrid(data=df_log_normal_melt, col='variable', col_wrap=3)
grid.map(plt.hist, "value", color="steelblue", cumulative = False)
```

```python
# WITH CUMMULATIVE
grid = sns.FacetGrid(data=df_log_normal_melt, col='variable', col_wrap=3)
grid.map(plt.hist, "value", color="steelblue", cumulative = True)
```
